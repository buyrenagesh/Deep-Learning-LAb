<p><span style="font-size: 14px;"><strong>for loop in python</strong></span></p>

```python
#range() used for printing statement multiple times using for loop

for i in range(10):
    print("Today, I'm enjoying the weather")
```

```python
# range() used to for finding cubes of number in a range using a for loop

for i in range(1,11):
    print(i**3)
```

<p><span style="font-size: 14px;"><strong>while loop in python</strong></span></p>

```python
#cubes can also be found using while statement

i=1
while i<11:
    print(i**3)
    i=i+1
```

<p><strong><span style="font-size: 20px;">Importing the libraries: time</span></strong></p>

```python
#Here this library accessed to calculate time is imported

import time
start_time= time.time()
for i in range(0,10001, 5):
    print(i**3)
end_time= time.time()
tm=end_time-start_time
print('\nThe time required is',tm)
```
